```python
#importacion de librerias
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

file = "Octubre_test_31 (3).xlsx"

data = pd.read_excel(file)
data = data.fillna(0)
def productividad(data):
    prod_pond = (data["Aprobadas"] * 5) + (data["Denegadas"] * 3) + (data["Llamadas Outbound"] * 2)
    return prod_pond
print("Introduce nombre de Agente:")
i = input()
data=data[data["Dia"] != 0]
data["prod_pond"] = productividad(data)
agente = data[data["Agente"] == i]
agente["date"] = pd.date_range(start="10-01-2016", end = "10-31-2016", freq='D')
agente = agente.set_index("date")


mes = agente[["Llamadas Outbound","Marcador","VoiceMail"]] 
mes.plot(kind = "bar",figsize=(15,5),stacked = (True),color=("y","b","orange"))
plt.show()




b = input()
data=data[data["Dia"] != 0]
data["prod_pond"] = productividad(data)
agente1 = data[data["Agente"] == b]
agente1["date"] = pd.date_range(start="10-01-2016", end = "10-31-2016", freq='D')
agente1 = agente1.set_index("date")

mes = agente1[["Llamadas Outbound","Marcador","VoiceMail"]] 
mes.plot(kind = "bar",figsize=(15,5),stacked = (True),color=("y","b","orange"))
plt.show()
d= {"agente":agente["prod_pond"],"agente1":agente1["prod_pond"]}
prod = pd.DataFrame(data=d)

prod.plot(figsize=(15,5))
plt.show()
```
